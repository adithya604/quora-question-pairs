```python
import pandas as pd
```

```python
model_A = pd.read_csv('29042017_1.csv', usecols=['is_duplicate'])
model_B = pd.read_csv('29042017_2.csv', usecols=['is_duplicate'])
merged = pd.concat([model_A, model_B], axis=1)
```

```python
final_model = pd.DataFrame()
final_model['is_duplicate'] = merged.mean(axis=1)
final_model.to_csv(
    path_or_buf='final.csv',
    index_label='test_id'
)
```
